```python
import pandas as pd
import matplotlib.pyplot as plt

# Customer Data
data = {
    "Customer_ID": [101, 102, 103, 104, 105, 106],
    "Name": ["Arun", "Priya", "Rahul", "Sneha", "Kiran", "Divya"],
    "Age": [22, 28, 35, 24, 31, 27],
    "Gender": ["Male", "Female", "Male", "Female", "Male", "Female"],
    "City": ["Chennai", "Madurai", "Coimbatore", "Chennai", "Salem", "Madurai"],
    "Income": [30000, 45000, 50000, 35000, 55000, 40000]
}

df = pd.DataFrame(data)

# Output 1
print("Customer Data")
print(df)

# Output 2
print("\nTotal Customers:", len(df))

# Output 3
print("Average Age:", round(df["Age"].mean(), 2))

# Output 4
print("Youngest Age:", df["Age"].min())

# Output 5
print("Oldest Age:", df["Age"].max())

# Output 6
print("Average Income:", round(df["Income"].mean(), 2))

# Output 7
print("Highest Income:", df["Income"].max())

# Output 8
print("Lowest Income:", df["Income"].min())

# Output 9
print("\nGender Count")
print(df["Gender"].value_counts())

# Output 10
print("\nCustomers by City")
print(df["City"].value_counts())

# Output 11
print("\nCustomers Above Average Income")
print(df[df["Income"] > df["Income"].mean()][["Name", "Income"]])

# Output 12
df["Age_Group"] = pd.cut(df["Age"], bins=[20, 25, 30, 35, 40],
                         labels=["21-25", "26-30", "31-35", "36-40"])
print("\nAge Group Distribution")
print(df["Age_Group"].value_counts())

# Output 13
print("\nMedian Income:", df["Income"].median())

# Output 14
print("Income Range:", df["Income"].max() - df["Income"].min())

# Output 15
print("Income Standard Deviation:", round(df["Income"].std(), 2))

# Bar Chart
plt.figure(figsize=(8,5))
plt.bar(df["Name"], df["Income"])
plt.title("Customer Income Analysis")
plt.xlabel("Customer Name")
plt.ylabel("Income")
plt.show()

# Pie Chart
plt.figure(figsize=(6,6))
df["Gender"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Gender Distribution")
plt.ylabel("")
plt.show()
```
